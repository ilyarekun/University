```python
import numpy as np
import math
np.random.seed(42)
```

```python
#task 1

matr = np.random.rand(640, 480)
matr_t = matr[:].T
print(matr, 4*"\n")
print(matr_t)
```

```python
#task 2

matr_2 = (matr_t * 255).astype(int)
matr_2[matr_2 < 50] = 0
print(matr_2)
```

```python
#if the task was to scale the mathix from any value range to [0,1]

matr_q = np.random.randint(-20, 20, (10, 10))
print(matr_q, 4*'\n')

min_val = np.min(matr_q)
max_val = np.max(matr_q)
matr_q = (matr_q - min_val) / (max_val - min_val)
print(matr_q)
```

```python
#task 3 (if it was ment to use Frobenius norm)

#1st way
matr_3 = matr_2[:]
frob_norm = np.linalg.norm(matr_3, ord = 'fro')
print(frob_norm, 4*'\n')
matr_3 = matr_3/frob_norm
print(np.linalg.norm(matr_3, 'fro'))
print(matr_3)
```

```python
# 2nd way
matr_4 = matr_2[:]
frob_norm = math.sqrt(np.sum(matr_4**2))
print(frob_norm, 4*'\n')
matr_4 = matr_4/frob_norm
print(math.sqrt(np.sum(matr_4**2)))
print(matr_4)
```
